```stata
***********************************************
*** Myoracer Mixed Effects regression
***********************************************
*init Stata to run locally
version 15
sysdir set OLDPLACE "\\stata.rz.uni-greifswald.de\StataSE$\ado\base\"
mata:mata clear
clear matrix
clear
set maxvar 32767
set more off
pause off
set excelxlsxlargefile on
sysdir set PERSONAL "~/brain/stata/ado/personal/"
sysdir set PLUS "~/brain/stata/ado/plus/"
discard
cd ~/brain/data/myo_stata



* load data

use strain_mapping.dta


* drop data without corresponding segments
drop if missing(value_t1) & missing(value)



* set followup var

gen followup = !baseline if !missing(baseline)

* drop unneccessary vars
drop index


*drop if missing(followup)
qui ds followup pat_id view aha_no side orientation value_t1 value, not
local covars "`r(varlist)'"
reshape wide `covars' value_t1 value, i(pat_id view aha_no side orientation) j(followup)

rename value_t10 t1_time0
rename value_t11 t1_time1
rename value0 t2_time0
rename value1 t2_time1
* gen baseline vals
g base_t1_time0 = t1_time0
g base_t2_time0 = t2_time0
g base_t1_time1 = t1_time0 if !( missing(t1_time1) | missing(t1_time0))
g base_t2_time1 = t2_time0 if !( missing(t2_time1) | missing(t2_time0))

* gen deltas
g delta_t1_time0 = 0 if !( missing(t1_time0))
g delta_t2_time0 = 0 if !( missing(t2_time0))
g delta_t1_time1 = t1_time1 - t1_time0 if !( missing(t1_time1) | missing(t1_time0))
g delta_t2_time1 = t2_time1 - t2_time0 if !( missing(t2_time1) | missing(t2_time0))

* gen quotients
g quot_t1_time0 = 1 if !( missing(t1_time0))
g quot_t2_time0 = 1 if !( missing(t2_time0))
g quot_t1_time1 = t1_time1 / t1_time0 if !( missing(t1_time1) | missing(t1_time0))
g quot_t2_time1 = t2_time1 / t2_time0 if !( missing(t2_time1) | missing(t2_time0))

# delimit ;
reshape long
delta_t1_time delta_t2_time
base_t1_time base_t2_time
quot_t1_time quot_t2_time
t1_time t2_time `covars', i(pat_id view aha_no side orientation) j(followup);
# delimit cr

* define interaction for EMB_res with recovery
g int_EMB_res_followup = EMB_res * followup

* setup different models
global MAPPING_VALUES "t1 t2"
global ORIENTATIONS "circumferential radial"
global VIEWS `"sax  lax"' //  "4cv" "2cv" "'
global SEGMENTWISE "global segmental"

di `"$VIEWS"'

*preserve

* LV only
drop if side != "LV"
*drop if !baseline

* drop missing EMB_res
drop if missing(int_EMB_res_followup)


* clear
local c=1 // counter for model save
local model_names ""
local cross_model_names ""
local fup_model_names ""
local model_rel_time ""

eststo clear

* cycle through
foreach map_val of global MAPPING_VALUES {
    foreach globallocal of global SEGMENTWISE {
        foreach view_name of global VIEWS {
            foreach orientation of global ORIENTATIONS {
                di "----------------------------------"
                di "Running Mixed Effects model for"
                di "map_val: `map_val'"
                di "orientation: `orientation'"
                di `"view: `view_name'"'
                di "globallocal: `globallocal'"
                di "----------------------------------"
                if "`globallocal'" == "global" {
                    local exclude_segs "aha_no == -1"
                    local group_level ""
                }
                if "`globallocal'" == "segmental" {
                    local exclude_segs "aha_no != -1"
                    local group_level "|| aha_no:"
                }
                
                local orientation_tag "`orientation'"
                
                if "`view_name'" == "sax" {
                    local view `"view == "sax""'
                    
                }
                if "`view_name'" == "lax" {
                    local view `"(view == "4cv" | view == "2cv")"'
                    
                    * rename strain in circ orienta to longit
                    if "`orientation'" == "circumferential" {
                        local orientation_tag "longitudinal"
                    }
                    
                } 

                
                capture drop base_relax_time
                capture drop delta_relax_time
                capture drop cross_relax_time
                clonevar base_relax_time = base_`map_val'_time
                clonevar delta_relax_time = delta_`map_val'_time
                clonevar cross_relax_time = `map_val'_time
                
                // rescale: consider changes in 10 ms
                replace base_relax_time = base_relax_time / 10
                replace delta_relax_time = delta_relax_time / 10
                replace cross_relax_time = cross_relax_time / 10
                
                if "`map_val'" == "t1" {
                    // rescale: consider changes in 100 ms
                    replace base_relax_time = base_relax_time / 10
                    replace delta_relax_time = delta_relax_time / 10
                    replace cross_relax_time = cross_relax_time / 10
                }
                
                label var base_relax_time "baseline relaxation time (%/ms)"
                label var delta_relax_time "delta relaxation time (%/ms)"
                label var cross_relax_time "relaxation time (%/ms)"
                

                
                
                
                /*
                di "Association of Baseline `map_val' with Recovery:";
                mixed value_strain c.base_`map_val'##i.followup  /* c.`map_val'#i.followup */
                `group_level'
                || pat_id:
                if `exclude_segs'
                & orientation == "`orientation'"
                & view == "`view'"
                ;
                 */
                di as err "Association of Delta `map_val' with Recovery:"
                
                
                // naive models for t1 time only (no diff for t2)
                if "`map_val'" == "t1" {
                    # delimit ;
                    
                    * effect modification by EMB result
                    eststo myocarditis_recovery`c', title(
                    "map_val: `map_val'
                    orientation: `orientation'
                    view: `view'
                    globallocal: `globallocal'"
                    ):
                    // run restricted model
                    qui mixed value_strain i.followup int_EMB_res_followup 1.EMB_res  /* c.`map_val'#i.followup */
                    `group_level'
                    || pat_id:
                    if `exclude_segs'
                    & orientation == "`orientation'"
                    & `view'
                    & aha_no != 16
                    ;
                    
                    * naive model with followup only
                    eststo naive`c', title(
                    "map_val: `map_val'
                    orientation: `orientation'
                    view: `view'
                    globallocal: `globallocal'"
                    ):
                    // run restricted model
                    qui mixed value_strain i.followup 1.EMB_res  /* c.`map_val'#i.followup */
                    `group_level'
                    || pat_id:
                    if `exclude_segs'
                    & orientation == "`orientation'"
                    & `view'
                    & aha_no != 16
                    ;



                    // do LR test for inclusion of nested model and store to ereturn matrix
                    lrtest naive`c' myocarditis_recovery`c';


                    // store to ereturn matrix
                    mat pval_lr = vecdiag(I(colsof(e(b)))) * `r(p)';
                    mat lr_chi2 = vecdiag(I(colsof(e(b)))) * `r(chi2)';
                    local colfulln "`:colfullnames e(b)'";
                    mat colnames pval_lr = `colfulln';
                    mat colnames lr_chi2 = `colfulln';
                    estadd matrix pval_lr = pval_lr, replace : naive`c';
                    estadd matrix lr_chi2 = lr_chi2, replace : naive`c';
                   
                    
                    local fup_model_names `"
                    `fup_model_names' 
                    naive`c' 
                    "`globallocal'
                     `orientation_tag' 
                     `view_name'"
                    "';
                     #delimit cr
                }
                
                # delimit ;
                
                * restricted model without interaction
                eststo restr`c', title(
                "map_val: `map_val'
                orientation: `orientation'
                view: `view'
                globallocal: `globallocal'"
                ):
                // run restricted model
                qui mixed value_strain c.base_relax_time i.followup i.EMB_res /* c.`map_val'#i.followup */
                `group_level'
                || pat_id:
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                
                * cross sectional model without interaction
                eststo cross`c', title(
                "map_val: `map_val'
                orientation: `orientation'
                view: `view'
                globallocal: `globallocal'"
                ):
                // run restricted model
                qui mixed value_strain c.cross_relax_time i.followup i.EMB_res /* c.`map_val'#i.followup */
                `group_level'
                || pat_id:
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                
                * cross sectional model with interaction, Effect modification by EMB_res?
                qui mixed value_strain c.cross_relax_time##i.EMB_res i.followup  /* c.`map_val'#i.followup */
                `group_level'
                || pat_id: 
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                // lr test for effect modification by EMB_res
                lrtest cross`c' .;
                
                
                // store to ereturn matrix
                    // store to ereturn matrix
                    mat pval_lr = vecdiag(I(colsof(e(b)))) * `r(p)';
                    mat lr_chi2 = vecdiag(I(colsof(e(b)))) * `r(chi2)';
                    local colfulln "`:colfullnames e(b)'";
                    mat colnames pval_lr = `colfulln';
                    mat colnames lr_chi2 = `colfulln';
                    estadd matrix pval_emb = pval_lr, replace : cross`c';
                    estadd matrix emb_chi2 = lr_chi2, replace : cross`c';
                
                
                capture drop int_base_followup;
                qui g int_base_followup = followup * base_relax_time;
                label var int_base_followup "Interaction followup x relaxation time (%/ms)";
                
                
                * prognostic value of baseline relax tme without EMB effect modification
                eststo prog`c', title(
                "map_val: `map_val'
                orientation: `orientation'
                view: `view'
                globallocal: `globallocal'"
                ):
                // run restricted model
                qui  mixed value_strain c.base_relax_time c.int_base_followup i.followup i.EMB_res /* c.`map_val'#i.followup */
                `group_level'
                || pat_id:
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                
                * baseline model with interaction, Effect modification by EMB_res?
                qui mixed value_strain c.base_relax_time c.int_base_followup c.int_base_followup#i.EMB_res i.EMB_res i.followup  /* c.`map_val'#i.followup */
                `group_level'
                || pat_id: 
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                
                lrtest prog`c' .;
                
                
                // store to ereturn matrix
                    // store to ereturn matrix
                    mat pval_lr = vecdiag(I(colsof(e(b)))) * `r(p)';
                    mat lr_chi2 = vecdiag(I(colsof(e(b)))) * `r(chi2)';
                    local colfulln "`:colfullnames e(b)'";
                    mat colnames pval_lr = `colfulln';
                    mat colnames lr_chi2 = `colfulln';
                    estadd matrix pval_emb = pval_lr, replace : prog`c';
                    estadd matrix emb_chi2 = lr_chi2, replace : prog`c';
                
                
                
                * full model with interaction
                qui mixed value_strain c.base_relax_time c.delta_relax_time i.followup  i.EMB_res /* c.`map_val'#i.followup */
                `group_level'
                || pat_id: 
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                
                // do lincom for delta and base = lc_1

                // do lincom for delta and followup var = lc_2
                qui xlincom (lc_1 = base_relax_time + delta_relax_time) (lc_2 = 1.followup + delta_relax_time), repost;  // repost to store into e
                
                // store estimates
                eststo full`c', title(
                "Relaxation time: `map_val'
                orientation: `orientation'
                view: `view_name'
                globallocal: `globallocal'"
                );
                
                // do LR test for inclusion of nested model and store to ereturn matrix
                lrtest full`c' restr`c';
                
                
                // store to ereturn matrix
                    // store to ereturn matrix
                    mat pval_lr = vecdiag(I(colsof(e(b)))) * `r(p)';
                    mat lr_chi2 = vecdiag(I(colsof(e(b)))) * `r(chi2)';
                    local colfulln "`:colfullnames e(b)'";
                    mat colnames pval_lr = `colfulln';
                    mat colnames lr_chi2 = `colfulln';
                    estadd matrix pval_lr = pval_lr, replace : full`c';
                    estadd matrix lr_chi2 = lr_chi2, replace : full`c';
                
                
                * full model with interaction, Effect modification by EMB_res?
                qui mixed value_strain c.base_relax_time c.delta_relax_time##i.EMB_res i.followup  /* c.`map_val'#i.followup */
                `group_level'
                || pat_id: 
                if `exclude_segs'
                & orientation == "`orientation'"
                & `view'
                & aha_no != 16
                ;
                
                // store estimates
                eststo full_emb`c', title(
                "Relaxation time: `map_val'
                orientation: `orientation'
                view: `view_name'
                globallocal: `globallocal'"
                );
                
                                // do LR test for inclusion of nested model and store to ereturn matrix
                lrtest full`c' full_emb`c';
                
                
                // store to ereturn matrix
                    // store to ereturn matrix
                    mat pval_lr = vecdiag(I(colsof(e(b)))) * `r(p)';
                    mat lr_chi2 = vecdiag(I(colsof(e(b)))) * `r(chi2)';
                    local colfulln "`:colfullnames e(b)'";
                    mat colnames pval_lr = `colfulln';
                    mat colnames lr_chi2 = `colfulln';
                    estadd matrix pval_emb = pval_lr, replace : full`c';
                    estadd matrix emb_chi2 = lr_chi2, replace : full`c';
                
                
                
                // collect info about model/label names
                local model_names `"
                `model_names' 
                full`c' 
                "`globallocal'
                 `orientation_tag' 
                 `view_name'"
                "';
                local cross_model_names `"
                `cross_model_names' 
                cross`c' 
                "`globallocal'
                 `orientation_tag' 
                 `view_name'"
                "';
                
                local model_rel_time `"
                `model_rel_time' 
                "`=strupper("`map_val'")'"
                "';
                
                di "finished `c++'";
                
                
                *eststo full`c++', addscalars(p_lr r(p), replace);
                /*
                # delimit ;
                mixe;
                                d value_strain c.`map_val'##i.followup /* c.`map_val'#i.followup */
                `group_level'
                /* || pat_id: */
                if `exclude_segs'
                & orientation == "`orientation'"
                & view == "`view'"
                ; */
                #delim cr

            }
        }

    }
}
```
